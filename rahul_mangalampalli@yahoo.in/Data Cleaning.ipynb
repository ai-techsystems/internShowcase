```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
```

```python
x = df.isnull().sum()
```

```python
y = dict(x)
y
```

```python
df['LotFrontage'].dropna().size
```

```python
df.dropna(subset = ['LotFrontage'],inplace =True)
```

```python
df.drop(['Alley','FireplaceQu','PoolArea','PoolQC','Fence','MiscFeature'],axis = 1,inplace = True)
```

```python
list1 = {}
for i in df.columns:
    if df[i].dtype == 'object':
        list1[i] = df[i].value_counts().index
list1    
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in df.columns:
    if df[i].dtypes == 'object':
        df[i].replace(np.nan,'Unlabelled',inplace = True)
        le.fit(df[i])
        df[i]= le.transform(df[i])
```

```python
df['GarageYrBlt'].replace(np.nan,df['GarageYrBlt'].mean(),inplace =True)
df['MasVnrArea'].replace(np.nan,df['MasVnrArea'].mean(),inplace =True)
```
