```python
"""
Created on Tue Aug 13 16:53:20 2019

Principal Component Analysis on the asteroids data
Perform a Principal Component Analysis on the asteroids data. 
What is the minimum number of principal components that contain 50% of the information? 
How many principal components for 90% information?

@author: Saransh
"""
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
pd.options.display.float_format = '{:,.2f}'.format
```

```python
#Read and process the data set
data = pd.read_csv('nasa.csv', header = 0)
data.head()
```

```python
#Removing the categorical value
data = data.drop(columns = ['Name','Neo Reference ID','Close Approach Date','Orbiting Body' ,'Orbit Determination Date', 'Equinox', 'Hazardous'])
data.head()
```

```python
#Standardize the data
X_std = StandardScaler().fit_transform(data.astype(np.float))
X_std[:3,:]
```

```python
#Find the variance each component holds
pca = PCA(n_components=None)
pca.fit(X_std.T)
explained_variance = pca.explained_variance_ratio_

# Create a data set where the second column is the number of components and the first column represents the percentage of inormation 
components_information =  pd.DataFrame(explained_variance*100,range(1, len(explained_variance) + 1), columns = ['Percentage_info'])

#components_information.index.name = 'Components'
components_information['Components']=components_information.index
#The correspnding information used by first 15 compnents
components_information[0:15]
```

```python
#The total number of components
components_information['Components'].shape[0]
```

```python
def no_of_components(arr, info_percent):
    """"
    Input: 
    arr - list containing inofrmation held by all components in percentage.
    info_percent - percentage of information reruired to represent the whole data
    
    Output/ Returns:
    the number of components required to represent the percentage of information provided as input
    """
    comps = 0
    sum_info = 0
    info_percent = float(info_percent)
    for i in range(len(arr)):
        if sum_info >= info_percent:
            break
        sum_info = sum_info + arr[i]
        comps += 1

    return comps

x = components_information.iloc[:,0].values
```

```python
#The minimum number of compenents that hold 50% of the information 
n50 = no_of_components(x,50)
n50
```

```python
#The total information in percentage stored by all 33 components
components_information['Percentage_info'].sum() 
```

```python
#The total information in percentage stored by all 33 components
components_information['Percentage_info'].sum() 
```

```python
#The minimum number of compenents that hold 90% of the information 
n90 = no_of_components(x,90)
n90
```
